#Data Understanding
Iris dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = pd.read_csv(url, names=names)
```

## Task 2: Data Visualization

- Univariate plots, visualisation of each individual feature for better understand.
- Multivariate plots, visualisation relationships between attributes.

Citations:
https://machinelearningmastery.com/visualize-machine-learning-data-python-pandas/

```python
dataset.hist()
dataset.plot(kind='density', subplots=True, layout=(2,2), sharex=False)
plt.show()
```

```python
dataset.plot(kind = 'density')
plt.show()
```

```python
import numpy as np
cor = dataset.corr()                              #obtain correlation matrix for dataset
#print cor
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cor, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,4,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(dataset)
plt.show()
```
